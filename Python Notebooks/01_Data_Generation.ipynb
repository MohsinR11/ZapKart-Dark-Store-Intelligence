```python
# ============================================================
# TABLE 5: fact_orders
# ~200,000 orders over 18 months
# This table is the heart of the entire project
# ============================================================

print("Generating fact_orders... this will take 1-2 minutes...")

start_date = datetime(2023, 1, 1)
end_date   = datetime(2024, 6, 30)
date_range = (end_date - start_date).days

# City to stores mapping
city_stores = dim_stores.groupby('city')['store_id'].apply(list).to_dict()
city_customers = dim_customers.groupby('city')['customer_id'].apply(list).to_dict()

# Hourly demand weights (quick commerce peaks at breakfast, lunch, evening)
hour_weights = {
    0: 0.3, 1: 0.2, 2: 0.1, 3: 0.1, 4: 0.1, 5: 0.2,
    6: 0.8, 7: 2.5, 8: 4.0, 9: 3.5, 10: 3.0, 11: 3.5,
    12: 5.0, 13: 4.5, 14: 3.0, 15: 2.5, 16: 2.8, 17: 3.5,
    18: 5.5, 19: 6.0, 20: 5.5, 21: 4.5, 22: 3.0, 23: 1.5
}

# Store-level order volume base (orders per day)
store_daily_orders = {
    'STR_001': 68, 'STR_002': 78, 'STR_003': 52,
    'STR_004': 82, 'STR_005': 71, 'STR_006': 91,
    'STR_007': 74, 'STR_008': 63, 'STR_009': 70, 'STR_010': 67
}

order_records = []
order_counter = 1

for day_offset in range(date_range):
    current_date = start_date + timedelta(days=day_offset)
    day_of_week  = current_date.weekday()  # 0=Mon, 6=Sun
    month        = current_date.month

    # Weekend multiplier
    day_multiplier = 1.35 if day_of_week >= 5 else 1.0
    # Festive season boost (Oct-Nov)
    season_multiplier = 1.45 if month in [10, 11] else (
                        1.20 if month in [12, 1]  else (
                        0.90 if month in [5, 6]   else 1.0))

    for store_id in store_ids:
        base_orders = store_daily_orders[store_id]
        daily_volume = int(base_orders * day_multiplier * season_multiplier
                           * np.random.uniform(0.82, 1.18))

        store_row  = dim_stores[dim_stores['store_id'] == store_id].iloc[0]
        store_city = store_row['city']
        store_lat  = store_row['latitude']
        store_lon  = store_row['longitude']

        store_customers = city_customers.get(store_city, [])

        for _ in range(daily_volume):
            # Pick hour based on demand weights
            hour = random.choices(list(hour_weights.keys()),
                                  weights=list(hour_weights.values()))[0]
            minute  = random.randint(0, 59)
            second  = random.randint(0, 59)
            order_time = current_date.replace(
                hour=hour, minute=minute, second=second)

            # Delivery time components (in minutes)
            pick_time     = max(2, np.random.normal(5.2, 1.8))
            pack_time     = max(1, np.random.normal(2.1, 0.7))
            dispatch_time = max(0.5, np.random.normal(1.5, 0.5))

            # Distance from store (0.3 to catchment radius)
            distance_km = round(np.random.uniform(0.3,
                                store_row['catchment_radius_km']), 2)

            # Delivery speed varies by store, time of day, distance
            base_speed   = 15  # km/h average
            peak_penalty = 1.4 if hour in [12, 13, 19, 20] else 1.0
            travel_time  = (distance_km / base_speed) * 60 * peak_penalty
            travel_time  = max(3, np.random.normal(travel_time, 1.5))

            total_delivery_minutes = (pick_time + pack_time +
                                      dispatch_time + travel_time)

            # Introduce realistic failures (some stores worse than others)
            failure_rate = {
                'STR_001': 0.18, 'STR_002': 0.12, 'STR_003': 0.28,
                'STR_004': 0.14, 'STR_005': 0.20, 'STR_006': 0.10,
                'STR_007': 0.22, 'STR_008': 0.25, 'STR_009': 0.16,
                'STR_010': 0.19
            }
            if random.random() < failure_rate[store_id]:
                total_delivery_minutes *= np.random.uniform(1.3, 2.2)

            sla_met = total_delivery_minutes <= 10

            # Payment & order value
            payment_method = np.random.choice(
                ['UPI', 'Card', 'COD', 'Wallet'],
                p=[0.52, 0.23, 0.15, 0.10])
            order_value = round(max(49,
                np.random.lognormal(5.6, 0.65)), 2)
            delivery_charge = 0 if order_value >= 149 else 25
            platform_fee    = round(order_value * 0.03, 2)
            discount_amount = round(order_value *
                np.random.choice([0, 0.05, 0.10, 0.15, 0.20],
                p=[0.45, 0.25, 0.15, 0.10, 0.05]), 2)
            net_order_value = round(
                order_value - discount_amount + delivery_charge + platform_fee, 2)

            # Delivery partner cost
            delivery_cost = round(15 + (distance_km * 8) +
                                  np.random.normal(0, 3), 2)

            # Order status
            status = np.random.choice(
                ['Delivered', 'Cancelled', 'RTO'],
                p=[0.91, 0.07, 0.02])

            order_records.append({
                'order_id':               f'ORD_{str(order_counter).zfill(8)}',
                'store_id':               store_id,
                'customer_id':            random.choice(store_customers) if store_customers else 'CUST_000001',
                'order_timestamp':        order_time,
                'order_date':             current_date.date(),
                'order_hour':             hour,
                'day_of_week':            current_date.strftime('%A'),
                'month':                  month,
                'quarter':                (month - 1) // 3 + 1,
                'is_weekend':             day_of_week >= 5,
                'payment_method':         payment_method,
                'order_value_inr':        order_value,
                'discount_amount_inr':    discount_amount,
                'delivery_charge_inr':    delivery_charge,
                'platform_fee_inr':       platform_fee,
                'net_order_value_inr':    net_order_value,
                'pick_time_mins':         round(pick_time, 2),
                'pack_time_mins':         round(pack_time, 2),
                'dispatch_time_mins':     round(dispatch_time, 2),
                'travel_time_mins':       round(travel_time, 2),
                'total_delivery_mins':    round(total_delivery_minutes, 2),
                'sla_met':                sla_met,
                'distance_km':            distance_km,
                'delivery_cost_inr':      delivery_cost,
                'order_status':           status,
                'num_items':              np.random.randint(1, 12),
                'had_substitution':       np.random.choice(
                                          [True, False], p=[0.12, 0.88]),
                'customer_rating':        np.random.choice(
                                          [1, 2, 3, 4, 5],
                                          p=[0.04, 0.06, 0.10, 0.35, 0.45])
                                          if status == 'Delivered' else None
            })
            order_counter += 1

fact_orders = pd.DataFrame(order_records)
fact_orders.to_csv(f"{RAW_PATH}fact_orders.csv", index=False)

print(f"✅ fact_orders generated: {fact_orders.shape[0]:,} rows")
print(f"   Date range: {fact_orders['order_date'].min()} to {fact_orders['order_date'].max()}")
print(f"   SLA Met %: {fact_orders['sla_met'].mean()*100:.1f}%")
print(f"   Avg delivery time: {fact_orders['total_delivery_mins'].mean():.1f} mins")
print(fact_orders.groupby('store_id')['order_id'].count())
```

```python
# ============================================================
# TABLE 6: fact_order_items
# Line items for every order
# ============================================================

print("Generating fact_order_items... this may take 2-3 minutes...")

sku_list      = dim_skus['sku_id'].tolist()
sku_price_map = dim_skus.set_index('sku_id')['unit_price_inr'].to_dict()
sku_cost_map  = dim_skus.set_index('sku_id')['cost_price_inr'].to_dict()
sku_cat_map   = dim_skus.set_index('sku_id')['category'].to_dict()

# Category popularity weights
cat_weights = {
    'Fruits & Vegetables': 0.22, 'Dairy & Eggs': 0.18,
    'Packaged Food': 0.20,       'Beverages': 0.12,
    'Personal Care': 0.10,       'Household': 0.09,
    'Baby & Kids': 0.04,         'Snacks & Confectionery': 0.05
}

skus_by_category = dim_skus.groupby('category')['sku_id'].apply(list).to_dict()

item_records = []
item_counter = 1

# Sample orders for item generation (use all orders)
for _, order in fact_orders.iterrows():
    n_items = order['num_items']

    # Pick categories for this order
    chosen_cats = random.choices(
        list(cat_weights.keys()),
        weights=list(cat_weights.values()),
        k=n_items
    )

    seen_skus = set()
    for cat in chosen_cats:
        available = skus_by_category.get(cat, sku_list)
        sku_id    = random.choice(available)

        # Avoid duplicate SKUs in same order
        attempts = 0
        while sku_id in seen_skus and attempts < 5:
            sku_id   = random.choice(available)
            attempts += 1
        seen_skus.add(sku_id)

        qty        = np.random.choice([1, 2, 3, 4],
                                       p=[0.60, 0.25, 0.10, 0.05])
        unit_price = sku_price_map.get(sku_id, 50)
        unit_cost  = sku_cost_map.get(sku_id, 40)
        item_total = round(unit_price * qty, 2)
        item_margin= round((unit_price - unit_cost) * qty, 2)

        item_records.append({
            'item_id':           f'ITM_{str(item_counter).zfill(9)}',
            'order_id':          order['order_id'],
            'store_id':          order['store_id'],
            'sku_id':            sku_id,
            'category':          cat,
            'quantity':          qty,
            'unit_price_inr':    unit_price,
            'unit_cost_inr':     unit_cost,
            'item_total_inr':    item_total,
            'item_margin_inr':   item_margin,
            'is_substituted':    False,
            'order_date':        order['order_date']
        })
        item_counter += 1

fact_order_items = pd.DataFrame(item_records)
fact_order_items.to_csv(f"{RAW_PATH}fact_order_items.csv", index=False)

print(f"✅ fact_order_items generated: {fact_order_items.shape[0]:,} rows")
print(fact_order_items.groupby('category')['item_id'].count())
```

```python
# ============================================================
# TABLE 7: fact_picker_activity
# ============================================================

print("Generating fact_picker_activity...")

store_pickers = dim_pickers.groupby('store_id')['picker_id'].apply(list).to_dict()

picker_records = []
for _, order in fact_orders[fact_orders['order_status'] == 'Delivered'].iterrows():
    store_id   = order['store_id']
    pickers    = store_pickers.get(store_id, [])
    if not pickers:
        continue

    picker_id  = random.choice(pickers)
    pick_start = order['order_timestamp']
    pick_end   = pick_start + timedelta(minutes=order['pick_time_mins'])

    # Experienced pickers pick faster
    picker_row  = dim_pickers[dim_pickers['picker_id'] == picker_id]
    skill       = picker_row['skill_level'].values[0] if len(picker_row) > 0 else 'Intermediate'
    efficiency  = {'Expert': 1.25, 'Intermediate': 1.00, 'Beginner': 0.78}.get(skill, 1.0)

    items_picked = order['num_items']
    pick_rate    = round((items_picked / order['pick_time_mins']) * efficiency, 3)

    picker_records.append({
        'activity_id':         f'ACT_{str(len(picker_records)+1).zfill(8)}',
        'order_id':            order['order_id'],
        'picker_id':           picker_id,
        'store_id':            store_id,
        'pick_start_time':     pick_start,
        'pick_end_time':       pick_end,
        'pick_duration_mins':  round(order['pick_time_mins'], 2),
        'items_picked':        items_picked,
        'pick_rate_per_min':   pick_rate,
        'errors_made':         np.random.choice([0, 1, 2],
                               p=[0.85, 0.12, 0.03]),
        'order_date':          order['order_date'],
        'shift':               'Morning' if order['order_hour'] < 12 else (
                               'Afternoon' if order['order_hour'] < 18 else 'Night')
    })

fact_picker_activity = pd.DataFrame(picker_records)
fact_picker_activity.to_csv(f"{RAW_PATH}fact_picker_activity.csv", index=False)
print(f"✅ fact_picker_activity generated: {fact_picker_activity.shape[0]:,} rows")
```

```python
# ============================================================
# TABLE 8: fact_substitutions
# ============================================================

print("Generating fact_substitutions...")

sub_orders = fact_orders[fact_orders['had_substitution'] == True]

sub_records = []
for _, order in sub_orders.iterrows():
    original_sku = random.choice(sku_list)
    substitute   = random.choice(sku_list)
    while substitute == original_sku:
        substitute = random.choice(sku_list)

    orig_cat  = sku_cat_map.get(original_sku, 'Packaged Food')
    sub_cat   = sku_cat_map.get(substitute,   'Packaged Food')

    # Same-category subs are accepted more
    same_cat  = orig_cat == sub_cat
    accepted  = np.random.choice(
        [True, False],
        p=[0.74, 0.26] if same_cat else [0.41, 0.59]
    )

    price_diff = round(
        sku_price_map.get(substitute, 50) - sku_price_map.get(original_sku, 50), 2
    )

    sub_records.append({
        'substitution_id':     f'SUB_{str(len(sub_records)+1).zfill(7)}',
        'order_id':            order['order_id'],
        'store_id':            order['store_id'],
        'original_sku_id':     original_sku,
        'substitute_sku_id':   substitute,
        'original_category':   orig_cat,
        'substitute_category': sub_cat,
        'same_category':       same_cat,
        'price_difference_inr':price_diff,
        'customer_accepted':   accepted,
        'reason_for_sub':      np.random.choice(
                               ['Out of Stock', 'Damaged', 'Expired'],
                               p=[0.78, 0.14, 0.08]),
        'order_date':          order['order_date']
    })

fact_substitutions = pd.DataFrame(sub_records)
fact_substitutions.to_csv(f"{RAW_PATH}fact_substitutions.csv", index=False)
print(f"✅ fact_substitutions generated: {fact_substitutions.shape[0]:,} rows")
```
